# Ejercicio 1

Realizar un programa que cumpla con las siguientes especificaciones:
* Solicitar el ingreso por teclado de 3 palabras. Cada una debe ingresarse de forma independiente.
* Determinar cual de las 3 palabras ingresadas tiene mayor longitud.


**Resolución**

```python
#Variables utilizadas
#Palab1: almacena la primera palabra ingresada por el usuario.
#Palab2: almacena la segunda palabra ingresada.
#Palab3: almacena la tercera palabra ingresada.
#Largo1: cantidad de letras de la primera palabra (Palab1).
#Largo2: cantidad de letras de la segunda palabra (Palab2).
#Largo3: cantidad de letras de la tercera palabra (Palab3).
```

```python
#Descripcion de instrucciones:
#El programa solicita al usuario ingresar tres palabras, una por una.
#Luego calcula la longitud (cantidad de letras) de cada palabra.
#Finalmente compara las longitudes para determinar cuál de las tres palabras es la más larga.
#Se muestra por pantalla cuál palabra tiene la mayor longitud.
```

```python
#Solicitar el ingreso de tres palabras
Palab1=input("Ingresa la primera palabra:")
Palab2=input("Ingresa la segunda palabra:")
Palab3=input("Ingresa la tercera palabra:")

#Calcular la longitud de cada palabra 
Largo1=len(Palab1)
Largo2=len(Palab2)
Largo3=len(Palab3)


#Determinar la palabra con mayor longitud 
if Largo1 >= Largo2 and Largo1 >= Largo3:
    print("La mayor longitud es de la palabra 1:", Palab1)

elif Largo2 >= Largo1 and Largo2 >= Largo3:
    print("La mayor longitud es de la palabra 2:", Palab2)

else: 
    print("La mayor longitud es de la palabra 3:", Palab3)
```

# Ejercicio 2

a)
Crear una función que reciba un número por parámetro y determine si el mismo es múltiplo de 3.
La función debe devolver verdadero si el número recibido es múltiplo de 3, o falso en caso contrario.
<br>
<br>
b)
Escribir un programa que solicite el ingreso de números positivos hasta recibir un cero. <br>
Utilizando la función creada en el punto anterior, evaluar si el número ingresado es múltiplo de 3.
Si lo es, guardar el número en una lista, si no lo es no realizar ninguna tarea.

Luego de finalizar la carga de números mostrar todos los números múltiplos de 3 guardados en la lista.

**Resolución**

```python
#Variables utilizadas
#NumIng: número ingresado por el usuario.
#ListaM: lista donde se guardan los múltiplos de 3.
```

Punto a

```python
#Descripción de instrucciones 
#Crea una función llamada `es_multiplo` que recibe un número como parámetro y devuelve `True` si el número es múltiplo de 3, o `False` si no lo es.
```

```python
# Función que verifica si un número es múltiplo de 3
def es_multiplo(NumIng):
    if NumIng % 3 == 0:
        return True
    else:
        return False
```

Punto b

```python
#Descripción de instrucciones:
#Se crea un programa que solicita al usuario ingresar números positivos.  
#Cuando el usuario ingresa 0, el programa termina.  
#Cada número ingresado se verifica con la función `es_multiplo`.  
#Si es múltiplo de 3, se guarda en la lista `ListaM`.
#Al final, se imprime la lista de múltiplos de 3.
```

```python
# Lista para guardar múltiplos de 3
ListaM = []

# Bucle para ingresar números
while True:
    NumIng = int(input("Ingresá un número positivo (0 para finalizar): "))
    if NumIng == 0:
        break
    if es_multiplo(NumIng):
        ListaM.append(NumIng)

# Mostrar los múltiplos de 3 guardados
print("Los múltiplos de 3 ingresados fueron:", ListaM)
```

# Ejercicio 3

a) Sin utilizar librerías externas escribir un programa que genere 100 valores para la siguiente función, considerando que el primer valor de variable x debe ser -50, y debe tener incremento de a 0.5.
Mostrar en pantalla cada valor de x con su correspondiente valor de f(x) generado

$f(x)= 3x^3 + 1,5x^2 + x$ + 0.35
<br>
<br>
b) Repetir el punto anterior utilizando Numpy
<br>
<br>
c) Graficar la función generada en el punto b (si no realiza el punto b, utilizar los datos generados en el punto a)

**Resolución**

```python
#Variables utilizadas:
#x_valores: lista con los valores de x generados manualmente.
#y_valores: lista con los resultados de f(x) para cada x.
#x_np: arreglo de x generado con numpy.
#y_np: arreglo de f(x) generado con numpy.
#x: valor actual de x usado en el bucle del punto a.
#incremento: cuánto aumenta x en cada paso.
#total_valores: cuántos valores se generan (100).
```

Punto a

```python
#Descripcion de instrucciones
#Se genera un total de 100 valores para x, empezando en -50 y sumando 0.5 en cada paso.
#Para cada valor de x se calcula f(x) = 3x^3 + 1.5x^2 + x + 0.35
#Se muestran en pantalla los pares (x, f(x)) y se guardan en listas.
```

```python
# Listas para guardar los valores
x_valores = []
y_valores = []

# Empezamos desde -50
x = -50

# Repetimos 100 veces
total_valores = 100
incremento = 0.5

for i in range(100):
    y = 3 * x**3 + 1.5 * x**2 + x + 0.35  # Fórmula dada
    x_valores.append(x)
    y_valores.append(y)
    print("x =", round(x, 2), "→ f(x) =", round(y, 2))
    x += incremento

```

Punto b

```python
#Descripcion de instrucciones: 
#Se repite lo anterior pero usando NumPy para crear el arreglo de x y calcular f(x) vectorizado.
```

```python
import numpy as np

# Generar los valores de x con numpy
desde = -50
hasta = desde + (total_valores - 1) * incremento
x_np = np.linspace(desde, hasta, total_valores) 

# Calcular f(x) con numpy
y_np = 3 * x_np**3 + 1.5 * x_np**2 + x_np + 0.35

# Mostrar los resultados
for i in range(total_valores):
    print("x =", round(x_np[i], 2), "→ f(x) =", round(y_np[i],2))
```

Punto c

```python
#Se utiliza matplotlib para graficar la función f(x) generada en el punto b.
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
plt.plot(x_np, y_np, color='blue', label='f(x)')
plt.title("Gráfico de f(x) = 3x^3 + 1.5x^2 + x + 0.35")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.legend()
plt.show()
```
